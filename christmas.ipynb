```python
import matplotlib.pyplot as plt
#plt.rcParams['font.family'] = 'Comic Sans MS'
import random
from matplotlib.gridspec import GridSpec
```

```python
def draw_neural_net(ax, left, right, bottom, top, layer_sizes):
    '''
    Draw a neural network cartoon using matplotilb.
    
    :usage:
        >>> fig = plt.figure(figsize=(12, 12))=
        >>> draw_neural_net(fig.gca(), .1, .9, .1, .9, [4, 7, 2])
    
    :parameters:
        - ax : matplotlib.axes.AxesSubplot
            The axes on which to plot the cartoon (get e.g. by plt.gca())
        - left : float
            The center of the leftmost node(s) will be placed here
        - right : float
            The center of the rightmost node(s) will be placed here
        - bottom : float
            The center of the bottommost node(s) will be placed here
        - top : float
            The center of the topmost node(s) will be placed here
        - layer_sizes : list of int
            List of layer sizes, including input and output dimensionality
    '''
    n_layers = len(layer_sizes)
    v_spacing = (top - bottom)/float(max(layer_sizes))
    h_spacing = (right - left)/float(len(layer_sizes) - 1)
    # Nodes
    for n, layer_size in enumerate(layer_sizes):
        layer_top = v_spacing*(layer_size - 1)/2. + (top + bottom)/2.
        for m in range(layer_size):
            circle = plt.Circle((n*h_spacing + left, layer_top - m*v_spacing), v_spacing/4.,
                                color='w', ec='k', zorder=4)
            ax.add_artist(circle)
    # Edges
    for n, (layer_size_a, layer_size_b) in enumerate(zip(layer_sizes[:-1], layer_sizes[1:])):
        layer_top_a = v_spacing*(layer_size_a - 1)/2. + (top + bottom)/2.
        layer_top_b = v_spacing*(layer_size_b - 1)/2. + (top + bottom)/2.
        for m in range(layer_size_a):
            for o in range(layer_size_b):
                line = plt.Line2D([n*h_spacing + left, (n + 1)*h_spacing + left],
                                  [layer_top_a - m*v_spacing, layer_top_b - o*v_spacing], c='g')
                ax.add_artist(line)


```

```python
def draw_neural_net_rot(ax, left, right, bottom, top, layer_sizes):
    '''
    Draw a neural network cartoon using matplotilb.
    
    :usage:
        >>> fig = plt.figure(figsize=(12, 12))
        >>> draw_neural_net(fig.gca(), .1, .9, .1, .9, [4, 7, 2])
    
    :parameters:
        - ax : matplotlib.axes.AxesSubplot
            The axes on which to plot the cartoon (get e.g. by plt.gca())
        - left : float
            The center of the leftmost node(s) will be placed here
        - right : float
            The center of the rightmost node(s) will be placed here
        - bottom : float
            The center of the bottommost node(s) will be placed here
        - top : float
            The center of the topmost node(s) will be placed here
        - layer_sizes : list of int
            List of layer sizes, including input and output dimensionality
    '''
    n_layers = len(layer_sizes)
    v_spacing = (top - bottom)/float(max(layer_sizes))
    h_spacing = (right - left)/float(len(layer_sizes) - 1)
    # Nodes
    for n, layer_size in enumerate(layer_sizes):
        layer_top = v_spacing*(layer_size - 1)/2. + (top + bottom)/2.
        for m in range(layer_size):
            color = random.choice(['r', 'b', 'goldenrod'])
            if n == 0 or n == 1:
                color = "brown"

            circle = plt.Circle((layer_top - m*v_spacing, n*h_spacing + left), v_spacing/4.,
                                color=color, ec='k', zorder=4)
            ax.add_artist(circle)
    # Edges
    for n, (layer_size_a, layer_size_b) in enumerate(zip(layer_sizes[:-1], layer_sizes[1:])):
        layer_top_a = v_spacing*(layer_size_a - 1)/2. + (top + bottom)/2.
        layer_top_b = v_spacing*(layer_size_b - 1)/2. + (top + bottom)/2.
        for m in range(layer_size_a):
            for o in range(layer_size_b):
                if n == 0:
                    lc = "brown"
                else :
                    lc = "g"
                line = plt.Line2D([layer_top_a - m*v_spacing, layer_top_b - o*v_spacing],
                                  [n*h_spacing + left, (n + 1)*h_spacing + left], c=lc, linewidth=5)
                ax.add_artist(line)
```

```python
fig = plt.figure(figsize=(12, 16))
gs = GridSpec(10, 1, figure=fig)
ax = fig.add_subplot(gs[:9, 0])
ax2 = fig.add_subplot(gs[9, 0])
ax2.axis('off')
ax.axis('off')
draw_neural_net_rot(ax, .1, .9, .1, .9, [2, 2, 7, 5, 4, 2, 1])

ax.set_title("Happy Christmas!", fontsize=30, fontstyle='italic')

ax.text(0.5, 0,"Not a descision tree!", fontsize=20, ha='center', va='center', fontstyle='italic')
string = "Wednesdays the 18"+r"$^\text{th}$"+" of December 2024 @ 5pm"
string2 = "Physics Common Room, Level 8"
string3 = "Tickets £10 Staff, £8 Students | From the Fundamental Admin Office (room 542)"
string4 = "Food, Entertainment, Raffle and DISCO!"
string_tot = string + "\n" + string2 + "\n" + string3 + "\n" + string4
ax2.text(0.5,0, string_tot, fontsize=20, ha='center', va='center', fontstyle='italic')

# fig.patch.set_facecolor("navajowhite")
# fig.patch.set_facecolor("powderblue")
# fig.patch.set_facecolor("mistyrose")
fig.patch.set_facecolor("lightcoral")
fig.tight_layout()

fig.savefig('nn.png')

# colour the nodes green and branches brown
# OR red baubles and green branches :D
```
